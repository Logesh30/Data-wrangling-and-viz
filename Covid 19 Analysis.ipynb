#  Filtering dataframe with only this columns ['continent','location','date','total_cases','total_deaths','gdp_per_capita','human_development_index']

# Data Cleaning 

# Removing all duplicates observations

```python
covid_upd.drop_duplicates
```

# Find missing values in all columns

```python
covid_upd.isnull()
```

# Remove all observations where continent column value is missing

```python
covid_upd.dropna(subset=['continent'],inplace=True)
```

# Fill all missing values with 0

```python
covid_upd.fillna(0,inplace=True)
covid_upd
```

# Date time format :

```python
covid_upd['date']=pd.to_datetime(covid_upd['date'])
```

```python
covid_upd['month']=covid_upd['date'].dt.month
covid_upd
```
